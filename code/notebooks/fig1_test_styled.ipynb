```python
import qutip as qt
import numpy as np
import matplotlib.pyplot as plt
from wigner_resolution.plotstyle import use_prl_style, save_figure, COLUMN_WIDTH

use_prl_style(use_tex=True)  # or False if you skipped LaTeX

psi = qt.coherent(50, 2.0)
xvec = np.linspace(-5, 5, 200)
W = qt.wigner(psi, xvec, xvec)

fig, ax = plt.subplots(figsize=(COLUMN_WIDTH, COLUMN_WIDTH))
cs = ax.contourf(xvec, xvec, W, 100, cmap='RdBu_r')
ax.set_xlabel(r'$x$')
ax.set_ylabel(r'$p$')
ax.set_aspect('equal')
ax.set_title(r'Coherent state $|\alpha = 2\rangle$')
plt.show()

save_figure(fig, "fig1_coherent_wigner")
```
