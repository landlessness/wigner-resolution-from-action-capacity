```python
import qutip as qt
import numpy as np
import matplotlib.pyplot as plt

# Quick sanity check: Wigner function of a coherent state
psi = qt.coherent(50, 2.0)
xvec = np.linspace(-5, 5, 200)
W = qt.wigner(psi, xvec, xvec)

fig, ax = plt.subplots(figsize=(4, 4))
ax.contourf(xvec, xvec, W, 100, cmap='RdBu_r')
ax.set_xlabel(r'$x$')
ax.set_ylabel(r'$p$')
ax.set_aspect('equal')
plt.show()
```
